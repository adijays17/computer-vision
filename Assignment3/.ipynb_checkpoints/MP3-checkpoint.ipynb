```python
img = cv2.imread("moon.bmp", 0)
```

```python
def get_histogram(image, bins):
    histogram = np.zeros(bins)
    for pixel in image:
        histogram[pixel] += 1
    return histogram

plt.plot(get_histogram(flat, 256))
```

```python
# create our cumulative sum function
def cumsum(a):
    a = iter(a)
    b = [next(a)]
    for i in a:
        b.append(b[-1] + i)
    return np.array(b)

# execute the fn
cs = cumsum(hist)

# display the result
plt.plot(cs)
```

```python
# re-normalize cumsum values to be between 0-255
nj = (cs - cs.min()) * 255
N = cs.max() - cs.min()
cs = nj / N
plt.plot(cs)
```

```python
# cast it back to uint8 since we can't use floating point values in images
cs = cs.astype('uint8')
plt.plot(cs)
```

```python
# get the value from cumulative sum for every index in flat, and set that as img_new
img_new = cs[flat]

# we see a much more evenly distributed histogram
plt.hist(img_new, bins=50)
```

```python
# put array back into original shape since we flattened it
img_new = np.reshape(img_new, img.shape)

# set up side-by-side image display
fig = plt.figure()
fig.set_figheight(15)
fig.set_figwidth(15)

fig.add_subplot(1,2,1)
plt.imshow(img, cmap='gray')

# display the new image
fig.add_subplot(1,2,2)
plt.imshow(img_new, cmap='gray')

plt.show(block=True)
cv2.imwrite("moon_hist_eq.bmp", img_new)
```

```python
#img = cv2.imread('new_image_hist.jpg', 0)
img = img_new
rows = img.shape[0]  #no. of rows in input image
cols = img.shape[1]  #no. of columns in input image
out = np.array([[0]*cols]*rows)  #initialize array to store LR image

#REGRESSION FOR LINEAR FIT:
x = []
y = []
for r in range(rows):
    for c in range(cols):
        x.append([1, r, c])
        y.append(img[r][c])

X = np.asarray(x)
X = X.T
y = np.asarray(y)
y.shape = (np.size(y), 1)

A = X.dot(X.T)
B = X.dot(y)

w = (np.linalg.pinv(A)).dot(B)

ynew = (X.T).dot(w)

#CREATE PLANE-FIT IMAGE:
ind = 0  #y index counter
for r in range(rows):
    for c in range(cols):
        out[r][c] = ynew[ind][0]
        ind += 1

#CREATE MEAN OF PLANE-FIT IMAGE:
out_mean = np.mean(out)
out_mean_img = np.array([[out_mean]*cols]*rows)

#CREATE LIGHTING CORRECTED IMAGE:
lc_img = (img - out) + out_mean_img
print(lc_img)
cv2.imwrite('lc_lin_histeq_moon.bmp', lc_img)
```

```python
#img = cv2.imread('new_image_hist.jpg', 0)
img = img_new
rows = img.shape[0]  #no. of rows in input image
cols = img.shape[1]  #no. of columns in input image
out = np.array([[0]*cols]*rows)  #initialize array to store LR image

#REGRESSION FOR QUADRATIC FIT:
x = []
y = []
for r in range(rows):
        for c in range(cols):
            x.append([1, r, c, r*r, c*c, r*c])
            y.append(img[r][c])

X = np.asarray(x)
X = X.T
y = np.asarray(y)
y.shape = (np.size(y), 1)

A = X.dot(X.T)
B = X.dot(y)

w = (np.linalg.pinv(A)).dot(B)

ynew = (X.T).dot(w)

#CREATE QUADRATIC SURFACE-FIT IMAGE:
ind = 0  #y index counter
for r in range(rows):
    for c in range(cols):
        out[r][c] = ynew[ind][0]
        ind += 1

#CREATE MEAN OF SURFACE-FIT IMAGE:
out_mean = np.mean(out)
out_mean_img = np.array([[out_mean]*cols]*rows)

#CREATE LIGHTING CORRECTED IMAGE:
lc_img = (img - out) + out_mean_img
print(lc_img)
cv2.imwrite('lc_qd_histeq_moon.bmp', lc_img)
```
