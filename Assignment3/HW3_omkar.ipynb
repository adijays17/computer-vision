# HW 3: Histogram Equalization

### I am using OpenCV and NumPy to access and work with the pixel values of the images.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

### create_hist function below creates a histogram depending on the specified number of bins and the input image.

```python
def create_hist(img_in, bins):
    hist = np.zeros(bins)
    
    for pixel in img_in:
        hist[pixel] += 1
    
    return hist
```

```python
def cumulative_sum(hist):
    hist = iter(hist)
    result = [next(hist)]
    
    for i in hist:
        result.append(result[-1] + i)
    
    return np.array(result)
```

```python
def normalize(c_sum):
    n = (c_sum - c_sum.min())*255
    d = c_sum.max() - c_sum.min()

    c_sum = n/d
    c_sum = c_sum.astype('uint8')
    
    return c_sum
```

```python
def image_recreate(result, new_image_path):
    image = (np.array(result)).astype(np.uint8)
    cv2.imwrite(new_image_path, result)
```

```python
def HistoEqualization(img_in):
    flattened_image = img_in.flatten()
    hist = create_hist(flattened_image, 256)
    plt.title("Initial histogram")
    plt.plot(hist)
    plt.show()
    c_sum = cumulative_sum(hist)
    c_sum = normalize(c_sum)
    new_image = c_sum[flattened_image]
    hist2 = create_hist(new_image, 256)
    plt.title("New Histogram")
    plt.plot(hist2)
    plt.show()
    new_image = np.reshape(new_image, img_in.shape)
    
    return new_image 
```

```python
img_in = cv2.imread("moon.bmp", 0)
plt.title("Initial Image")
plt.imshow(img_in, cmap='gray')
plt.show()
```

```python
new_image = HistoEqualization(img_in)
image_recreate(new_image, "new_image_hist.jpg")
plt.title("Image after histogram equalization")
plt.imshow(new_image, cmap='gray')
plt.show()
```

### Lighting correction

```python
def lighting_linear(img_in):
    tmp_A = []
    tmp_b = []
    
    for i in range(len(img_in)):
        for j in range(len(img_in[0])):
            tmp_A.append([i, j , 1])
            tmp_b.append(img_in[i][j])
    b = np.matrix(tmp_b).T
    A = np.matrix(tmp_A)
    fit = (A.T * A).I * A.T * b

    n_image = []
    for i in range(len(img_in)):
        n_image.append([])
        for j in range(len(img_in[0])):
            z = fit.item(0)*i + fit.item(1)*j + fit.item(2)
            n_image[i].append(z)
    
    mean = np.mean(n_image)
    #print(mean)
            
    image = n_image-img_in + mean
    
    return image
```

```python
def lighting_quad(img_in):
    x = []
    y = []
    for i in range(len(img_in)):
        for j in range(len(img_in[0])):
            x.append([i, j, i*i, j*j, i*j, 1])
            y.append(img_in[i][j])
            
    b = np.matrix(y).T
    A = np.matrix(x)
    fit = (A.T * A).I * A.T * b

    n_image = []
    for i in range(len(img_in)):
        n_image.append([])
        for j in range(len(img_in[0])):
            z = fit.item(0)*i + fit.item(1)*j + fit.item(2)*(i*i) + fit.item(3)*(j*j) + fit.item(4)*(i*j) + fit.item(5)
            n_image[i].append(z)
    
    mean = np.mean(n_image)
    image = n_image-img_in + mean
    
    return image

```

```python
def scaling(image):
    return 255*(image - np.min(image))/np.ptp(image).astype(int)

def truncation(image):
    for i in range(len(image)):
        for j in range(len(image)):
            if image[i][j] < 0:
                image[i][j] = 0
            elif image[i][j] > 255:
                image[i][j] = 255
    return image
```

```python
image = lighting_linear(new_image)
image = truncation(image)
```

```python
image_recreate(image, "new_moon_linear.jpg")
plt.title("Image after linear lighting correction")
plt.imshow(image, cmap='gray')
```

```python
image = lighting_quad(new_image)
image = truncation(image)
image_recreate(image, "new_moon_quad.jpg")
plt.title("Image after quadratic lighting correction")
plt.imshow(image, cmap='gray')
```
